# Popularity Based System 

```python
#here I'm adding to two dataFrame "ratings" and "isbn" where I'm assiging the new Dataframe called "Rating_with_NAMES
Ratings_with_name = ratings.merge(books, on = 'isbn')
#The on='isbn' parameter specifies that the merging should be done based on the 'isbn' column in both dataframes.
#This means that rows with the same 'isbn' value in both dataframes will be merged.
Ratings_with_name.head()
```

```python
#seeking the groupby name and counting here ratings that a single book has recived..........
num_rating_df = Ratings_with_name.groupby('book_title').count()['rating'].reset_index()
# The reset_index method helps in resetting the index to the default integer-based index or another specified index.
num_rating_df.rename(columns ={'rating': 'num_rating'} , inplace = True)
num_rating_df
```

```python
#want to see that how each book has recived the avg_ratings  
avg_rating_df = Ratings_with_name.groupby('book_title')['rating'].mean().reset_index()
avg_rating_df = avg_rating_df.rename(columns={'rating': 'avg_rating'})
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df, on = 'book_title')
#mergig is happening here on the basis of book_title 
popularity_df
```

```python
popularity_df[popularity_df['num_rating']>=250].sort_values('avg_rating',ascending = False).head()
```

```python
# here I'm droping the duplicats from the book title and adding the book title , book_author, num_rating
popular_df = popularity_df.merge(books, on = 'book_title').drop_duplicates('book_title')[['book_title','book_author','num_rating','avg_rating']]
popular_df
```

# Filtering Process

```python
#Users those who have not given rating more than 200
Ratings_with_name.groupby('user_id').count()['rating']>200
```

```python
x = Ratings_with_name.groupby('user_id').count()['rating']>200
#This line filters the users based on the boolean condition. The expression x[x] selects only the True values from the boolean Series. 
#The resulting Series contains only those users who have more than 200 ratings.
user_1 = x[x].index
```

```python
#Ratings_with_name['user_id']:- This extracts the 'user_id' column from the Ratings_with_name DataFrame, creating a Pandas Series.
filtered_rating = Ratings_with_name[Ratings_with_name['user_id'].isin(users_1)]
```

```python
y = filtered_rating.groupby('book_title').count()['rating']>=50
famous_books= y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['book_title'].isin(famous_books)]
#So, in summary, the purpose of isin in this context is to filter the DataFrame to include only those rows where the 'user_id' is found 
# in the set of user_ids specified by users_1
```

```python
pt = final_ratings.pivot_table(index ='book_title',columns= 'user_id', values = 'rating')
```

```python
pt.fillna(0, inplace = True)
```

```python
pt.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score =cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(pt.index== book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x:x[1], reverse = True)[1:6] 
   
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Message in a Bottle')
```
